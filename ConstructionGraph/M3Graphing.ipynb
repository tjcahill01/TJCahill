```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df_const = pd.read_csv('ConstructionTimeSeriesDatav2.csv') 
```

```python
fig,ax = plt.subplots()
ax.plot(df_const['Month-Year'],df_const['Public Construction'],color='k')
ax.plot(df_const['Month-Year'],df_const['Private Construction'],color='b')
fig.suptitle('Private & Public Construction Spending by Month', fontsize = 16)

#TimeSeries Data Labels
ax.text(146,46647,'Private',fontsize=13,verticalalignment='center',horizontalalignment='left', color='b')
ax.text(146,16721,'Public',fontsize=13,verticalalignment='center',horizontalalignment='left')

#Formatting the Plot Space
ax.xaxis.set_tick_params(bottom = False)
ax.yaxis.set_tick_params(left = False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylim(0,max(df_const['Total Construction'])+50) #Add buffer space 

#Label for y-axis & x-axis
ax.set_ylabel('Construction Spending',fontsize='12')
ax.set_xlabel('Year',fontsize='12')

#Tick Mark Labels
ax.set_xticks(df_const['Month-Year'][::12])
ax.set_xticklabels(df_const['Month-Year'][::12])
ax.set_xticklabels(['{:.0f}'.format(2002+x/12) for x in ax.get_xticks()])
ax.set_yticklabels(['{:,.0f}'.format(y) for y in ax.get_yticks()])   #Adds a comma to the y-axis quantities

#Highlighted Area Formatting & Label
ax.axvspan(71,90,ymin=0,ymax = 300,facecolor='red',alpha=0.2)    
ax.text(81,90000,'Great''\n''Recession',fontsize=12,verticalalignment='center',horizontalalignment='center')

fig.set_size_inches(9,5)
```

During the Great Recession, private contruction abruptly declined year-over-year, while public contruction continued its steady upward trajectory.
